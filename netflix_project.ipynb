## Un-nesting


```python
# Unnesting CASTs
constraint=df['cast'].apply(lambda x: str(x).split(', ')).tolist()
df_cast=pd.DataFrame(constraint,index=df['title'])
df_cast=df_cast.stack()
df_cast=pd.DataFrame(df_cast)
df_cast.reset_index(inplace=True)
df_cast=df_cast[['title',0]]
df_cast.columns=['title','cast']
df_cast
```

```python
# Unnesting DIRECTORS
constraint=df['director'].apply(lambda x: str(x).split(', ')).tolist()
df_director=pd.DataFrame(constraint,index=df['title'])
df_director=df_director.stack()
df_director=pd.DataFrame(df_director)
df_director.reset_index(inplace=True)
df_director=df_director[['title',0]]
df_director.columns=['title','director']
df_director
```

```python
# Unnesting COUNTRYS
constraint=df['country'].apply(lambda x: str(x).split(', ')).tolist()
df_country=pd.DataFrame(constraint,index=df['title'])
df_country=df_country.stack()
df_country=pd.DataFrame(df_country)
df_country.reset_index(inplace=True)
df_country=df_country[['title',0]]
df_country.columns=['title','country']
df_country
```

```python
# Unnesting Listed IN
constraint=df['listed_in'].apply(lambda x: str(x).split(', ')).tolist()
df_listed=pd.DataFrame(constraint,index=df['title'])
df_listed=df_listed.stack()
df_listed=pd.DataFrame(df_listed)
df_listed.reset_index(inplace=True)
df_listed=df_listed[['title',0]]
df_listed.columns=['title','listed_in']
df_listed
```

```python
# merging all table with the original table to get final Unnested table
df_1=df_cast.merge(df_country, on='title',how='outer')
df_2=df_1.merge(df_director,on='title',how='outer')
df_3=df_2.merge(df_listed, on='title', how='outer')
df_4= df[['title','show_id','type','month_added','month_name_added','year_added','release_year','rating','duration',]]  # getting values from original table
df_5=df_3.merge(df_4, on='title', how='outer')    # merging with original table
df_unnested= df_5[['show_id','type','title','director','cast','country','month_added','month_name_added',	'year_added','release_year','rating','duration','listed_in']]  # Re-arranging the sequence of columns as per original table.
df_unnested    # Here I have dropped 'Discription' column for simplicity.
```

```python
# Ungrouping on the basis of type and duration for Movies
constraint=df['duration'].apply(lambda x: str(x).split(' ')).tolist()
df_duration=pd.DataFrame(constraint,index=df['title'])

df_duration
df_6=df[['title','type']]
df_7=df_duration.merge(df_6, on='title',how='outer')
df_movies= df_7.loc[df_7['type']== 'Movie']
df_movies
```

```python
# Ungrouping on the basis of type and duration for TVShows
constraint=df['duration'].apply(lambda x: str(x).split(' ')).tolist()
df_duration=pd.DataFrame(constraint,index=df['title'])
df_duration.columns=['duration','listed_in']
df_duration
df_6=df[['title','type']]
df_7=df_duration.merge(df_6, on='title',how='outer')
df_TVshow= df_7.loc[df_7['type']== 'TV Show']
df_TVshow
```

# Question_5 (Missing Value & Outlier check )

```python
#Calculating the missing data
for i in df.columns:
    null_rate = df[i].isna().sum()/len(df) * 100
    if null_rate > 0 :
        print("{} null rate: {}%".format(i,round(null_rate,2)))
```

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['cast'].replace(np.nan,'No data',inplace=True)
df['director'].replace(np.nan,'No data',inplace=True)
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.head()
# Here we will be replacing the missing country with the most country (mode), cast, and director with no data.
```

```python
# Outlier Check
sns.set(rc = {'figure.figsize':(15,10)})
sns.boxplot(y='release_year', data=df_unnested)
plt.title('Box Plot of Release Year')
plt.show()
```

# Question_7 (Insights and recommendations)

1) Netflix has a significant amount of missing data for some of the columns, especially for the 'director' and 'cast' columns, which could limit the insights that can be drawn from the data.

2) The dataset has more movies than TV shows, indicating that Netflix may be more focused on producing movies.

3) The US, India, United Kingdom, Canada, and France are the top countries producing content on Netflix, suggesting that Netflix is focusing on producing content in these countries to cater to the local audience.

4) The growth in content on Netflix started in 2013, with a significant increase in the number of movies in 2018 and 2019.

5) The growth in content on Netflix is higher in the first three months and the last three months of the year, indicating that Netflix may be releasing more content during the holiday season.

6) There is a negative relationship between drama and documentary movies, suggesting that Netflix may be producing more drama movies for independent and international films, and more documentary movies for mainstream films.

7) The majority of the movies on Netflix have a duration ranging from 85 min to 120 min, indicating that this is the most popular duration for movies on the platforme

8) Increase focus on producing content in countries where there is high demand for it. For example, in India, where there is a high demand for Bollywood movies and TV shows, Netflix can invest in producing more local content.

9) There is an increase in the content of both Movies and TV Shows 2015 onwards.
